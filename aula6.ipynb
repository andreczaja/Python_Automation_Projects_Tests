```python
import pandas as pd
df = pd.read_csv('wine_quality_classification.csv').head(30)
```

```python
df
```

```python
df['novo_indice'] = [x for x in range(10,40)]
df
```

```python
df.set_index('novo_indice',drop=True, inplace=True)
df
```

```python
df['alcohol_classification'] = df['alcohol'].apply(lambda x: 'weak' if x < 10 and x > 8 else ('moderate' if x <= 12 else 'strong'))
```

```python
df_agrupado = df.set_index(['quality_label','alcohol_classification'])
```

```python
df_agrupado = df_agrupado.sort_index(level='quality_label',ascending=True)
```

```python
# df_agrupado.index.names = ['quality_label', 'alcohol_classification']
df_agrupado.xs('moderate',level='alcohol_classification')[['fixed_acidity']]
```

```python
df_agrupado.reset_index(inplace=True)
df_agrupado
```
