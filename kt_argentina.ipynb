```python
import pandas as pd
```

```python
df_data = pd.read_csv('atividades_maio.csv',sep= ',', index_col=0)
df_data
```

```python
df_data.reset_index(inplace=True)
df_data
```

```python
df_data['date'] = pd.to_datetime(df_data['date'])
df_data
```

```python
df_data = df_data.sort_values(by=['task_name', 'date'])
df_data
```

```python
status_dict = {
    'status 1': 1,
    'status 2': 2,
    'status 3': 3,
    'status 4': 4,
    'status 5': 5
}
df_data['status_num'] = df_data['status'].map(status_dict)
df_data
```

```python
from datetime import date
all_dates = pd.date_range(start=df_data['date'].min(), end=date.today())
```

```python
all_tasks = df_data['task_name'].unique()
```

```python
full_index = pd.MultiIndex.from_product([all_dates, all_tasks], names=['date', 'task_name'])
```

```python
df_data
```

```python
df_data = df_data.groupby(['date', 'task_name'])['status_num'].max().reset_index()
```

```python
df_data
```

```python
df_data = df_data.set_index(['date', 'task_name'])
df_data
```

```python
df_data = df_data[['status_num']].reindex(full_index).groupby('task_name').ffill().reset_index()
```

```python
df_data['status_num'].fillna(1,inplace=True)
df_data
```
