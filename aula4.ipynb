```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({'A': [1,2,np.nan,4,np.nan],
                   'B': [5,np.nan,np.nan,6,np.nan],
                   'C': [1,2,3,6,np.nan],
                   'D': [2,4,7,10,np.nan],
                   }
)
df
```

```python
df.dropna(axis=1, thresh=3)
```

```python
df.fillna(0)
```

```python
df.ffill()
```

```python
for i in df.columns:
    df[i].fillna(value=df[i].mean(),inplace=True)
df
```
