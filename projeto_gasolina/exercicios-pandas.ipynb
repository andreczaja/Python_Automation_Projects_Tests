### 4. Você deve ter percebido que as colunas DATA INICIAL e DATA FINAL estão formatadas como string. Utilizando o método pd.to_datetime(), converta ambas colunas para Timestamp / Datetime.

```python
df_consolidado['DATA INICIAL'] = pd.to_datetime(df_consolidado['DATA INICIAL'])
df_consolidado['DATA FINAL'] = pd.to_datetime(df_consolidado['DATA FINAL'])
```

### 5. Crie uma nova coluna para representar o ano e o mês(aaaa-mm), utilizando a coluna DATA FINAL como referência. 

```python
df_consolidado['DATA FINAL_ANO_MES'] = df_consolidado['DATA FINAL'].apply(lambda x: str(x.year) + '-' + str(x.month))
df_consolidado
```

### 6. Utilizando o value_counts(), liste todos os tipos de produtos contidos na base de dados.

```python
df_tipoproduto = pd.DataFrame(df_consolidado['PRODUTO'].value_counts())
df_tipoproduto
```

```python
df_tipoproduto.drop('count',axis=1,inplace=True)
df_tipoproduto.reset_index()
```

### 7. Filtre o DataFrame para obter apenas dados da 'GASOLINA COMUM'. Grave em uma nova variável.

```python
df_gas_comum = df_consolidado[df_consolidado['PRODUTO'] == 'GASOLINA COMUM']
df_gas_comum
```

### 8. Qual o preço médio de revenda da gasolina em agosto de 2008?

```python
df_gas_comum[df_gas_comum['DATA FINAL_ANO_MES'] == '2008-8'].groupby('DATA FINAL_ANO_MES')[['PREÇO MÉDIO REVENDA']].mean()
```

### 9. Qual o preço médio de revenda da gasolina em maio de 2014 em São Paulo?

```python
df_sp_maio = df_gas_comum[(df_gas_comum['DATA FINAL_ANO_MES'] == '2014-5') & (df_gas_comum['ESTADO'] == 'SAO PAULO')].groupby(['DATA FINAL_ANO_MES','ESTADO'])[['PREÇO MÉDIO REVENDA']].mean()
df_sp_maio
```

### 10. Você conseguiria descobrir em qual(quais) estado(s) a gasolina ultrapassou a barreira dos R$ 5,00? E quando isso ocorreu?

```python
df_gas_maior_5 = df_gas_comum[df_gas_comum['PREÇO MÉDIO REVENDA'] > 5]
df_gas_maior_5 = df_gas_maior_5[['ESTADO','PREÇO MÉDIO REVENDA','DATA FINAL_ANO_MES']]
df_gas_maior_5 = df_gas_maior_5.groupby(['ESTADO','DATA FINAL_ANO_MES'])[['PREÇO MÉDIO REVENDA']].mean()
df_gas_maior_5
```

### 11. Qual a média de preço dos estados da região sul em 2012?

```python
df_sul_2012 = df_consolidado[(df_consolidado['REGIÃO']=='SUL') & (df_consolidado['DATA FINAL_ANO_MES'].str.contains('2012',case=False,na=False))].groupby('ESTADO')[['PREÇO MÉDIO REVENDA']].mean().round(2) # fight
df_sul_2012
```

### 12. Você conseguiria obter uma tabela contendo a variação percentual ano a ano para o estado do Rio de Janeiro?

```python
import locale
locale.setlocale(locale.LC_TIME, 'pt_BR.UTF-8')
df_rj = df_consolidado[df_consolidado['ESTADO']=='RIO DE JANEIRO']
df_rj['nome_mes'] = df_rj['DATA FINAL'].dt.month_name(locale='pt_BR.UTF-8')
df_rj
```

```python
df_rj['PRODUTO'] = df_rj['PRODUTO'].str.replace('OLEO', 'ÓLEO', regex=True)

df_rj['ANO'] = df_rj['DATA FINAL_ANO_MES'].str[:4]
df_rj = df_rj.groupby(['PRODUTO','ANO'])[['PREÇO MÉDIO REVENDA']].mean().pct_change().fillna(1).round(2) # fight
# df_rj =  df_rj.groupby('PRODUTO')[['PREÇO MÉDIO REVENDA']]
df_rj
```

```python
df_rj.rename(columns={'PREÇO MÉDIO REVENDA':'VARIAÇÃO PERCENTUAL'},inplace=True)
df_rj.fillna(0)
```
