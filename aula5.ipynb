```python
import pandas as pd
data = {'Classe': ['Júnior','Júnior','Pleno','Pleno','Sênior','Sênior'],
        'Nome': ['Jorge','Carlos','Roberta','Patrícia','Bruno','Vera'],
        'Venda' :[200,120,340,124,243,350]}
```

```python
df = pd.DataFrame(data)
df
```

```python
df.sort_values(by='Nome')
```

```python
df.groupby('Classe')[['Venda']].mean().sort_values(by='Venda',ascending=False)
```

```python
df.drop('Classe',axis=1)
```

```python
df2 = df.copy()
df2['Venda'] = [150,432,190,230,410,155]
df['Mês'] = 'Janeiro'
df['NúmeroMesAno'] = 12025
df2['Mês'] = 'Fevereiro'
df2['NúmeroMesAno'] = 22025
df2
```

```python
df
```

```python
df3 = pd.concat([df, df2])
df3
```

```python
df3.sort_values(by=['NúmeroMesAno','Venda'],inplace=True)
df3
```

```python
df3_agrupado = df3.groupby(['Classe','Mês','NúmeroMesAno'])[['Venda']].sum().sort_values(by='NúmeroMesAno').reset_index(level='NúmeroMesAno',drop=True)
df3_agrupado
```
